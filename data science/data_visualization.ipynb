```python
#visualize the survival rate with age of passengers
```

```python
survived_age=df.groupby('age')['survived'].mean()
survived_age
```

```python
plt.plot(survived_age.index,survived_age.values)
plt.title('Survival Rate by age!')
plt.xlabel('age')
plt.ylabel('survived_age')

plt.show()
```
